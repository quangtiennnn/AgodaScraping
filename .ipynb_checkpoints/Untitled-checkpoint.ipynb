```python
hotel = df[['hotel_id', 'hotel_name', 'addressline1', 'city', 'state','longitude','latitude','url', 'numberrooms', 'yearopened', 'yearrenovated', 'number_of_reviews', 'rating_average', 'star_rating','accommodation_type']]
hotel.head(3)
```

```python
hotel.shape
```

```python
# Step 1: Split the accommodation types by ',' and expand to individual entries
hotel['accommodation_type'] = hotel['accommodation_type'].apply(lambda x: [item.lower() for item in x.split(', ')] if isinstance(x, str) else [])

# Step 2: Use MultiLabelBinarizer to handle the one-hot encoding for lists of categories
from sklearn.preprocessing import MultiLabelBinarizer

# Initialize the binarizer
mlb = MultiLabelBinarizer()

# Fit and transform the split accommodation types
one_hot_encoded = mlb.fit_transform(hotel['accommodation_type'])

# Create a DataFrame with the one-hot encoded columns
one_hot_df = pd.DataFrame(one_hot_encoded, columns=mlb.classes_)

# Concatenate with the original DataFrame (dropping the original list column for clarity)
df_final = pd.concat([hotel.drop(columns=['accommodation_type']), one_hot_df], axis=1)
```

```python
import requests

def check_url_status(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36"
    }
    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 404:
            return 0  # URL not accessible (404)
        else:
            return 1  # URL accessible
    except requests.exceptions.RequestException:
        return 0  # Return 0 if there is any other exception (e.g., network error)

# Example usage
url = sample.url.values[0]
status = check_url_status(url)
print(status)
```

```python
from urllib.parse import urlparse, urlunparse

# Original URL
def url_edited(url):
    try:
        parsed_url = urlparse(url)
        new_path = "/vi-vn" + parsed_url.path
        return urlunparse(parsed_url._replace(path=new_path, query=""))
    except:
        return []
```
